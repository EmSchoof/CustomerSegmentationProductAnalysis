### Dataset description
**Source**: Google Analytics hit (pageview) level data, by date and by article title.

**Field descriptions**:
- Date - year, month, day of hit.
- Page Title - article title that was viewed by user
- Age - age of user that visited the site
- Gender - gender of user that visited the site
- Source/Medium - the referral source of the hit or the website that the user was on before visiting whowhatwear.com
- Pageviews - a hit of a url on our site that is being tracked by theAnalytics tracking code.
- Unique Pageviews - represents the number of sessions during which that page was viewed one or more times.

## Question 3: What are the channels that are driving engagement? Which traffic sources should we lean into, which traffic sources should we deprioritize?

In order to assess which channels drive engagement, the top columns within the dataset provided would be: *Date, Source/Medium, Pageviews, and Unique Pageviews*. Additionally, since the core audience demographic has been identified, the success and lapse in success of these channels can be compared and contrasted between the age and gender demographics. However, For the sake of this assessment, only the top audience demographic, **women ages 25-34**, will be analyzed in order to answer the questions above.

```python
# Import necessary modules
import pandas as pd
import numpy as np

# Import data
%store -r combo_content_df
print(combo_content_df.shape)
combo_content_df.head()
```

*Assessment of Who What Wear's Core Audience (Group 1)'s Main Engagement Channels*

```python
# Groupby Source / Medium
engagement_df = combo_content_df.drop(columns='PageTitle').groupby(['Source / Medium']).sum().sort_values(
                                        by='Unique Pageviews', ascending=False).reset_index()
engagement_df
```

*Assess Total Unique Pageviews of Who What Wear Articles in Dataset*

```python
# Create a general function for percent
def calculate_percentage(val, total):
    return round((val / total)*100, 3)
```

```python
# Define total Unique Pageviews of Who What Wear Articles in Dataset
total_unique_views = engagement_df['Unique Pageviews'].sum()
print(total_unique_views)
```

```python
engagement_df.iloc[1]['Source / Medium']
```

```python
# Calculate percentage of all 38 Source / Mediums
percent_unique_pageviews = []
for idx, source in engagement_df['Source / Medium'].iteritems():
    unique_views = engagement_df.iloc[idx]['Unique Pageviews']
    percent_uniq = round(calculate_percentage(unique_views, total_unique_views))
    percent_unique_pageviews.append(percent_uniq)
```

```python
# Add percent unique as column
engagement_df['Percent of Total Unique Pageviews'] = pd.Series(percent_unique_pageviews)
engagement_df.head(10)
```

```python
# Separate out top 10 Source / Mediums
sources_pageviews = list(engagement_df['Unique Pageviews'])
len(sources_pageviews)
```

*Visualize*

```python
# Import necessary modules
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10,10)})
```

```python
# Define a custom palette
positive_color_palette = sns.cubehelix_palette(10, start=2, rot=0, dark=0, light=.95, reverse=True)
negative_color_palette = sns.cubehelix_palette(10, reverse=True)
```

```python
# Plot counts and show percentages
ax = sns.barplot(x='Source / Medium', y='Unique Pageviews',
                 data=engagement_df[0:10], palette=positive_color_palette)

# Percentages added from source code:
# https://stackoverflow.com/questions/31749448/how-to-add-percentages-on-top-of-bars-in-seaborn/31754317
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2.,
            height + 5,
            '{:1.2f}'.format( (height/sum(engagement_df['Unique Pageviews'])) * 100),
            ha="center")  
ax.set(title='Top Percent of Unique Pageviews to WWW site by Medium [for Core Audience]', ylabel='Percent')
plt.xticks(rotation=90)
plt.savefig('topsources.png', dpi=300)
plt.show()
```

```python
# Plot counts and show percentages
ax = sns.barplot(x='Source / Medium', y='Unique Pageviews',
                 data=engagement_df[29:39], palette=negative_color_palette)

# Percentages added from source code:
# https://stackoverflow.com/questions/31749448/how-to-add-percentages-on-top-of-bars-in-seaborn/31754317
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2.,
            height + 5,
            '{:1.2f}'.format( (height/sum(engagement_df['Unique Pageviews'])) * 100),
            ha="center")  
ax.set(title='Bottom Percent of Unique Pageviews to WWW site by Medium [for Core Audience]', ylabel='Percent')
plt.xticks(rotation=90)
plt.savefig('bottomsources.png', dpi=300)
plt.show()
```
